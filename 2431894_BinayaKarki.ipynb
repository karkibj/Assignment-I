# Outlier Ditection Task3

***Ditecting outliers***

Calculating IQR for the score

```python
# Calculate IQR for Score
Q1_score = southAsia['score'].quantile(0.25)
Q3_score = southAsia['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score
lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

# Calculate IQR for GDP per Capita
Q1_gdp = southAsia['Log GDP per capita'].quantile(0.25)
Q3_gdp = southAsia['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp
lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

# Identify outliers
outliers = southAsia[
    (southAsia['score'] < lower_bound_score) | (southAsia['score'] > upper_bound_score) |
    (southAsia['Log GDP per capita'] < lower_bound_gdp) | (southAsia['Log GDP per capita'] > upper_bound_gdp)
]

print("Outlier Countries based on Score and GDP per Capita:")
outliers
```


Simply IQR refers to inter quartile range which provides the difference between Q3_score and Q1_score.In simple words it is us to measure of statistical dispersion, which is the spread of the data

 ***Scatter plot***

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot for Score vs GDP with outliers highlighted
plt.figure(figsize=(10, 7))
sns.scatterplot(data=southAsia, x='Log GDP per capita', y='score', label='Countries', color='blue')
sns.scatterplot(data=outliers, x='Log GDP per capita', y='score', label='Outliers', color='red')
plt.title('Scatter Plot: Score vs GDP per Capita (Outliers Highlighted)')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Score')
plt.legend()
plt.show()
```

as we can see how outliers is far from the other data points , these kinds of data might effect on overall view of the dataframe. For instance if we are willing to draw the linear line for the given data it will effect on the shape of the line

```python
# Detect outliers for both regions
def detect_outliers(data, column):
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return data[(data[column] < lower_bound) | (data[column] > upper_bound)]

south_asia_outliers = detect_outliers(southAsia, 'score')
middle_east_outliers = detect_outliers(middle_east, 'score')

print("South Asia Outliers:")
print(south_asia_outliers)
print("\nMiddle East Outliers:")
print(middle_east_outliers)

# Scatter plot to visualize outliers
plt.figure(figsize=(10, 6))
sns.scatterplot(x='score', y='Log GDP per capita', data=southAsia, label='South Asia', color='blue')
sns.scatterplot(x='score', y='Log GDP per capita', data=middle_east, label='Middle East', color='green')
plt.scatter(south_asia_outliers['score'], south_asia_outliers['Log GDP per capita'], color='red', label='Outliers South Asia')
plt.scatter(middle_east_outliers['score'], middle_east_outliers['Log GDP per capita'], color='orange', label='Outliers Middle East')
plt.title("Outliers in Score and GDP per Capita")
plt.xlabel("Happiness Score")
plt.ylabel("GDP per Capita")
plt.legend()
plt.show()
```
